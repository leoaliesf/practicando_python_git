```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset('tips')
tips.head(2)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip',hue='sex',marginal_ticks=True, marginal_kws=dict( fill = True, multiple= 'stack'))#kind='hist' dentro kws bins=25
plt.show()
```

```python
sns.pairplot(data=tips, hue='time')
plt.show()
```
