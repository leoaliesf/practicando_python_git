```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark',font="verdana",font_scale=1) #style= diseño, palette= deep, muted, font_scale tamaño, font tipo de letra
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
tip= sns.load_dataset('tips')
tip
```

```python
sns.displot(data=tip, x='total_bill')
```

```python
sns.displot(data=tip, x='total_bill', y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill',hue='sex',kind='kde',legend=False, alpha=0.35)
plt.title('grafica')
plt.show()
```

```python
sns.histplot(data=tip, x='tip',bins=18,cumulative=False, hue='sex',stat='frequency',multiple='stack')#multiple= layer, ddodge, fill stat= probability percent density default count
plt.show()
```

```python
sns.kdeplot(data=tip,x='tip',hue='sex', fill=True, bw_adjust=1) #fill devuelve el relleno
plt.show()
```

```python
sns.ecdfplot(data=tip,x='tip',hue='sex', stat='proportion')
plt.show()
```

```python
sns.displot(data=tip,x='tip',hue='sex',kind='hist', multiple='stack')#kind = hist
plt.show()
```

```python
sns.countplot(data=tip , x= 'day',hue= 'sex')
plt.show()
```

```python
sns.stripplot(data=tip , x= 'day',y='tip',hue= 'sex', dodge=True)
plt.show()
```

```python
sns.swarmplot(data=tip , x= 'day',y='tip',hue= 'sex', dodge=True)
plt.show()
```

```python
sns.boxplot(data=tip , x= 'day',y='tip',hue= 'sex', dodge=True, palette='pastel')
sns.swarmplot(data=tip , x= 'day',y='tip',hue= 'sex', dodge=True)
plt.show()
```

```python
sns.violinplot(data=tip , x= 'day',y='tip',hue= 'sex', dodge=True, split='True')
plt.show()
```

```python
sns.catplot(data=tip , x= 'day',y='tip',hue= 'sex', dodge=True, col='time')
plt.show()
```

```python
sns.pairplot(data=tip,hue='day')
plt.show() 
```

```python
markers = {"Male": 'v', "Female": 'o'}
sns.scatterplot(data=tip, x='total_bill', y='tip', hue='time',style='sex', markers=markers) 
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
plt.show()
```

```python
sns.relplot(data= tip, x= 'total_bill', y = 'tip',  hue= 'day', style= 'time', size='size')#, kind= 'line'
plt.show()
```
