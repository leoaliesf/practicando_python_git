```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark',font="verdana",font_scale=1) #style= diseño, palette= deep, muted, font_scale tamaño, font tipo de letra
sns.barplot(x=['A','B','C'], y=[1,3,2])
plt.show()
```

```python
tip= sns.load_dataset('tips')
tip
```

```python
sns.displot(data=tip, x='total_bill')
```

```python
sns.displot(data=tip, x='total_bill', y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill',hue='sex',kind='kde',legend=False, alpha=0.35)
plt.title('grafica')
plt.show()
```
