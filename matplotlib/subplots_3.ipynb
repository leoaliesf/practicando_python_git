```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,25)# crea una cadena de valores de 
y= np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3) # no es necesario colocar nrows y ncols
axes[(1,1)].plot(x,y, 'r')
fig.tight_layout() #separa un poco mas las graficas
```

```python
y2 = np.cos(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,6))#figsize su nombre lo indica para cambiar el tamño de la figura
axes[0].plot(x,y,label="$sin(x)$")#lo que aparece en la leyenda
#axes[0].plot(x,y,'r')
axes[0].legend()#para que aparezca la legenda

axes[0].set_title('grafica seno')
axes[0].set_xlabel('X')
axes[0].set_ylabel('y')

axes[1].set_ylabel('y')
axes[1].set_title('grafica coseno')
axes[1].plot(x,y2,label="$cos(x)$")
axes[1].legend(loc='lower left')
#axes[1].plot(x,y2)
fig.tight_layout
```
