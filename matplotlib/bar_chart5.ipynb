```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.linspace(2,12,5)
x=np.array(['1980','1990','2000','2010','2020'])
plt.style.use('seaborn-v0_8-dark-palette')
```

```python
plt.bar(x,y, label="aqui tanbien", width=0.6, alpha=0.6)
plt.plot(x,y, 'r--', label="Soy yo", linewidth= 2)
plt.title('modificando tabla')
plt.xlabel('titulo eje x')
plt.ylabel('titulo eje y')
plt.legend(loc='upper center')
plt.show()
```

```python
plt.barh(x,y, label="aqui tanbien", alpha=0.6)
plt.plot(y,x, 'r--', label="Soy yo", linewidth= 2)
plt.title('modificando tabla')
plt.xlabel('titulo eje x')
plt.ylabel('titulo eje y')
plt.legend(loc='upper center')
plt.show()
```
