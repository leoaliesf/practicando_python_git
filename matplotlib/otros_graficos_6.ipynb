```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(0,50,100)
```

```python
plt.hist(data,bins=15, histtype='step')
plt.show
```

```python
data=np.append(data,150)# ingreso de un dato anormal, notch da la forma de convergencia en la figura
plt.boxplot(data, vert=False, patch_artist=True, notch=True,showfliers=True)#showfliers muestra los datos fuera de lo usual
plt.show() #vert voltea la figura habitual
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N))**2
colors= np.random.rand(N)
```
