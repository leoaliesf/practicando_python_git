```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,30)
y = x ** (1/2)
```

```python

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.6])
axes.plot(x,y, 'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.6])
axes2 = fig.add_axes([0.3,0.16,0.2,0.3])
axes.plot(x,y, 'b')
axes2.plot(y,x,'r--')
axes2.set_facecolor('white')
fig.show()
```
