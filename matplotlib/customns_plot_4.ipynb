```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,30)
```

```python
print(plt.style.available)
```

```python
plt.style.use('fast')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+1,'r--')
ax.plot(x, x+2,'b:')
ax.plot(x, x+3,'g.-')
ax.plot(x, x+4,'mx:')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, x+1, color = 'red', alpha = 0.3) # alpha transparencia en la linea, color = https://matplotlib.org/3.5.1/tutorials/colors/colors.html
ax.plot(x, x+2,color = 'blue',linewidth= 8) # linewidth grosor de la linea
ax.plot(x, x+3,color = '#2CA252', linestyle= '--', marker='v') # linestyle como quieres tu linea
ax.plot(x, x+4,color = '#AAA657', marker = 'p')#marker colocar en tus puntos alguna figura (p,o,v,x)
```
