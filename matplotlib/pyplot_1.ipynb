```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,15)
y = x**2
x
```

```python
plt.plot(x,y) #plt.plot(x,y, 'X') x= b= blue, r = red, y= yellow
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'gD:')
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python
x = np.linspace(0,10,10)
y = x **2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.hist(y,color='r')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.hist(y,color='r')
plt.show()
```
